```python
import re
import pandas as pd
```

```python
base_url = 'https://books.toscrape.com/'
book_details = []
for i in b:
    book_url = base_url + i.h3.a['href']
    print(book_url)
    b_request = requests.get(book_url)
    data = BeautifulSoup(b_request.text,'html.parser')
    data_book = data.find(class_ = 'product_page')
    title =data_book.h1.string
    price = data_book.find(class_ = 'price_color').string
    quantity = list(data_book.find(class_ = 'instock availability').stripped_strings)[0]
    quantity = re.search('\d+',quantity).group()
    price = re.search('[\d.]+',price).group()
    book_details.append([title,book_url,price,quantity])
    df = pd.DataFrame(book_details,columns = {'Title','Link','Price','Quantity in stock'} )
```

```python
all_url = ['https://books.toscrape.com/catalogue/page-1.html']
current_page = 'https://books.toscrape.com/catalogue/page-1.html'
base_url = 'https://books.toscrape.com/catalogue/'
response = requests.get(current_page)
book_details = []


while response.status_code == 200 :
    data = BeautifulSoup(response.text,'html.parser')
    b = data.find_all(class_ = 'product_pod')
    base_url_books = 'https://books.toscrape.com/catalogue/'
    next_page = data.find(class_ = 'next')
    if(next_page is None):
        break
    for i in b:
        book_url = base_url_books + i.h3.a['href']
#         print(book_url)
        b_request = requests.get(book_url)
#         print(b_request.status_code)
        data = BeautifulSoup(b_request.text,'html.parser')
        data_book = data.find(class_ = 'product_page')
#         print(data_book)
        title =data_book.h1.string
        price = data_book.find(class_ = 'price_color').string
        quantity = list(data_book.find(class_ = 'instock availability').stripped_strings)[0]
        quantity = re.search('\d+',quantity).group()
        price = re.search('[\d.]+',price).group()
        book_details.append([title,book_url,price,quantity])
        df = pd.DataFrame(book_details,columns = {'Title','Link','Price','Quantity in stock'} )

    
    
    next_url =base_url + next_page.a['href']
#     print(next_url)
    all_url.append(next_url)
    current_page = next_url
    response = requests.get(current_page)
```

```python
df.to_csv('allbooks.csv',index = False)
```
